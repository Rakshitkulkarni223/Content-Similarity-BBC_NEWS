```python
from facebook_scraper import get_posts
import datetime
import json

class DateTimeEncoder(json.JSONEncoder):
    def default(self, z):
        if isinstance(z, datetime.datetime):
            return (str(z))
        else:
            return super().default(z)
        
json_data_file =open('HindiDataset.json','w',encoding='utf-8')

json_data_file.write('[')
```

```python
for post in get_posts('BBCnewsHindi', pages=10, cookies='cookies.txt',options={"comments": True, "reactors": True,"progress":True,"posts_per_page": 10}):

    post_comments={}
    post_comments["text"]=post["text"]
    post_comments["comments_full"]=post['comments_full']
    print(len(post['comments_full']))
    print(" Done!! \n\n")



    caption_comments_and_replies = {}

    comments=[]

    comments_and_replies=[]

    caption_comments_and_replies['content']=post_comments['text']

    all_comments = post_comments['comments_full']
    length_all_comments=len(all_comments)

    comments_and_replies_list=[]



    for pos in range(length_all_comments):

        comments_and_replies_dict={}


        comments_and_replies_dict['comment'] = all_comments[pos]['comment_text']
        comments_and_replies_dict['commenter_name'] = all_comments[pos]['commenter_name']
        
        comments_and_replies_dict['related'] = 1

        all_replies = all_comments[pos]['replies']
        length_all_replies = len(all_replies)

        replies=[]

        for repl_pos in range(length_all_replies):
            dict_reply = dict()
            dict_reply[all_replies[repl_pos]['comment_text']] = 1
            replies.append(dict_reply)

        comments_and_replies_dict['replies'] = replies

        comments_and_replies_list.append(comments_and_replies_dict)

    caption_comments_and_replies['comments_and_replies']=comments_and_replies_list
    

    json.dump(caption_comments_and_replies, json_data_file,ensure_ascii=False)
    json_data_file.write(',')


print("Completed!!!")
json_data_file.write(']')
json_data_file.close()
```
